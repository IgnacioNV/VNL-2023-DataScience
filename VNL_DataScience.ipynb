```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
vnl = pd.read_csv("https://raw.githubusercontent.com/IgnacioNV/VNL-2023-DataScience/main/VNL2023.csv")
vnl
```

```python
liberos = vnl[vnl["Position"] == "L"]
liberos
```

```python
promRecepcionLiberos = liberos["Receive"].mean()
mejoresLReceptores = liberos[liberos["Receive"] >= promRecepcionLiberos]
mejorLReceptor = mejoresLReceptores.iloc[0]
mejorLReceptor
```

```python
# Obtener los datos del DataFrame
recepcionLiberos = liberos["Receive"]
nombresLiberos = liberos["Player"]

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Crear la gráfica de barras
ax.stem(recepcionLiberos, nombresLiberos)

# Configurar las etiquetas y el título
ax.set_xlabel('Receive')
ax.set_ylabel('Player')
ax.set_title('Recepciones por Jugador')

# Mostrar la gráfica
plt.show()
```

```python
jugadoresArgentinos = vnl[vnl["Country"] == "Argentina"]
jugadoresArgentinos
```

```python
ataqueArgentina = jugadoresArgentinos["Attack"]
ataqueArgentina
```

```python
fig, ax = plt.subplots()
ax.stem(ataqueArgentina, jugadoresArgentinos["Player"])
ax.set_xlabel('Attack')
ax.set_ylabel('Player')
ax.set_title('Ataques por Jugador')
plt.show()
```

```python
maxAtaqueArgentino = jugadoresArgentinos["Attack"].max()
maxAtacanteArgentino = jugadoresArgentinos["Player"][jugadoresArgentinos["Attack"] == maxAtaqueArgentino]
maxAtacanteArgentino
```

```python
ordenadosPorAtaque = jugadoresArgentinos.sort_values(by="Attack", ascending=False)
ordenadosPorAtaque
```

```python
ordenadosPorRecepcion = jugadoresArgentinos.sort_values(by="Receive", ascending=False)
ordenadosPorRecepcion
```

```python
armadoresVNL :pd.DataFrame = vnl[vnl["Position"] == "S"]
mejoresArmadores = armadoresVNL.sort_values(by="Set", ascending = False)
mejorArmador : pd.Series = mejoresArmadores.iloc[0]
nombreMejorArmador = mejorArmador["Player"]
nombreMejorArmador
```

```python
puntasVNL = vnl[vnl["Position"] == "OH"]
mejoresPuntaAtacantes = puntasVNL.sort_values(by="Attack", ascending = False)
mejorPuntaAtacante = mejoresPuntaAtacantes.iloc[0]
nombreMejorPuntaAtacante = mejorPuntaAtacante["Player"]
nombreMejorPuntaAtacante
```

```python
mejoresPuntaReceptores = puntasVNL.sort_values(by= "Receive", ascending = False)
mejorPuntaReceptor = mejoresPuntaReceptores.iloc[0]
nombreMejorPuntaReceptor = mejorPuntaReceptor["Player"]
nombreMejorPuntaReceptor
```

```python
mejoresLReceptores = liberos.sort_values(by="Receive", ascending = False)
mejorLReceptor = mejoresLReceptores.iloc[0]
nombreMejorLReceptor = mejorLReceptor["Player"]
nombreMejorLReceptor
```

```python
mejoresLDefensores = liberos.sort_values(by="Dig", ascending = False)
mejorLDefensor = mejoresLDefensores.iloc[0]
nombreMejorLDefensor = mejorLDefensor["Player"]
nombreMejorLDefensor
```

```python
centralesVNL = vnl[vnl["Position"] == "MB"]
mejoresCentralesAtacantes = centralesVNL.sort_values(by="Attack", ascending = False)
mejorCentralAtacante = mejoresCentralesAtacantes.iloc[0]
nombreMejorCentralAtacante = mejorCentralAtacante["Player"]
nombreMejorCentralAtacante
```

```python
mejoresCentralesBloqueadores = centralesVNL.sort_values(by="Block", ascending = False)
mejorCentralBloqueador = mejoresCentralesBloqueadores.iloc[0]
nombreMejorCentralBloqueador = mejorCentralBloqueador["Player"]
nombreMejorCentralBloqueador
```
